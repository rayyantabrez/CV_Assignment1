```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# Read the original image
img = cv2.imread(r'C:\Users\saive\OneDrive\Desktop\WhatsApp Image 2024-01-27 at 11.00.31_027f9a7a.jpg')  
# converting because opencv uses BGR as default
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# converting to gray scale
gray = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)

# remove noise
img_blur = cv2.GaussianBlur(gray, (3, 3), 0)

# convolute with sobel kernels
sobelx = cv2.Sobel(img_blur, cv2.CV_64F, 1, 0, ksize=5)  # x
sobely = cv2.Sobel(img_blur, cv2.CV_64F, 0, 1, ksize=5)  # y

# Plotting images
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(RGB_img)
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(sobelx, cmap='gray')
plt.title("Sobel-x Edge Detection")

plt.subplot(1, 3, 3)
plt.imshow(sobely, cmap='gray')
plt.title("Sobel-y Edge Detection")

plt.show()
```

```python
sobel_combined = cv2.magnitude(sobelx, sobely)
plt.imshow(sobel_combined, cmap='gray')
plt.title("Combined Sobel Edge Detection")
plt.show()


```
